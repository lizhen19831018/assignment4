```python
#Chapter 11 Exercise 1
import re
import sys

#Define the function
def matchWord(word):
    try:
        #Open the file
        fhand = open(r'C:\Users\Dennis Li\mbox.txt', 'r')
    except:
        print('File cannot be opened.')
        sys.exit(0)
        
    count = 0
    #Search the word
    for line in fhand:
        line = line.rstrip()
        if re.search(word, line):
            count += 1
    fhand.close()
    
    #Print the result
    print('mbox.txt had %d lines that matched %s.' % (count, word))

#Call the function
matchWord(input('Enter a regular expression: '))
```

```python
#Chapter 11 Exercise 2
import re
import sys

#Define the function
def visionAverage(file):
    try:
        #Open the file
        fhand = open(file, 'r')
    except:
        print('File cannot be opened.')
        sys.exit(0)
        
    newRvs = []
    #Find new revision
    for line in fhand:
        line = line.rstrip()
        x = re.findall('^New Revision: ([0-9.]+)', line)
        for vision in x:
            newRvs.append(int(vision))
    fhand.close()
    try:
        #print the result
        print(sum(newRvs)/len(newRvs))
    except:
        print('New revision is not found.')
        
#Call the function
visionAverage(input('Enter file: '))
```

```python
#Call the function
visionAverage(input('Enter file: '))
```

```python
#Chapter 12 Exercise 2
import re
import socket

#Input the URL
url = input('Enter URL: ')
#Get the host
host = None

if re.search('^(?:http(s)?:\/\/)', url):
    host = url.split('/')[2]
else:
    host = url.split('/')[0]

#Read the URL
count = 0
try:
    mysock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    mysock.connect((host, 80))
    cmd = ('GET %s HTTP/1.0\r\n\r\n' % url).encode()
    mysock.send(cmd)

    while True:
        data = mysock.recv(512)
        if len(data) < 1 or count >= 3000:
            break
        print(data.decode()[0: (3000 - count)],end='')
        count += len(data.decode()[0: (3000 - count)])
except:
    print('Cannot open the URL.')

mysock.close()

print('\r\nTotal number of characters is %d' %count)
```

```python
#Chapter 12 Exercise 4

import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl

#Define the function
def countPagraph(url):
    # Ignore SSL certificate errors
    ctx = ssl.create_default_context()
    ctx.check_hostname = False
    ctx.verify_mode = ssl.CERT_NONE

    html = urllib.request.urlopen(url, context=ctx).read()
    soup = BeautifulSoup(html, 'html.parser')

    # Retrieve all of the anchor tags
    tags = soup('p')
    print(len(tags))

#Call the function
countPagraph(input('Enter - '))
```

```python
#Call the function
countPagraph(input('Enter - '))
```
